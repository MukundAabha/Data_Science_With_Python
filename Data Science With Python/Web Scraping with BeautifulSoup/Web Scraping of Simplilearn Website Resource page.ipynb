#### 1: View and print the Simplilearn webpage content in a good format

```python
#Import the required libraries
from bs4 import BeautifulSoup
#import urllib2 ; insted of this excute the below command
from urllib.request import urlopen
```

```python
#Provide the url for web scraping
url='https://www.simplilearn.com/resources'
```

```python
#Scrape the web page document
#webpage = urllib2.urlopen(url) #Change this step to below one
webpage = urlopen(url)
```

```python
#Create the soup object and parse the webpage using an html parser
s1_soup = BeautifulSoup(webpage,'html.parser')
```

```python
#Close the webpage object
webpage.close()
```

```python
#View the soup contents
s1_soup.contents
```

```python
#Use the prettify() method to view the well-formatted html output
print(s1_soup.prettify())
```

#### 2: View the title

```python
#View the title of the scraped document
s1_soup.head
```

```python
#Find the title of the scraped document
s1_soup.title
```

#### 3: Print all the href links present in the Simplilearn webpage

```python
#Find all href links
for href in s1_soup.findAll('a',href=True):
    print(href['href'])
```

#### 4: Search and print the resource headers of the Simplilearn webpage 

```python
#View all the headers name
for article in s1_soup.findAll("h2"):
    print(article.string)
```

```python
#the div information is obtained from the web page (view page source)
for article_topic in s1_soup.findAll("h4"):
    print(article_topic.string)
```
