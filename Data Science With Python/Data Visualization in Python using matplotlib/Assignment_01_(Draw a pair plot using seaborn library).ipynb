#### 2: Write a user-defined function for origin  


```python
#1-USA 2-Europe ,3-Asia
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'
#use apply function
df_auto_datset['origin'] = df_auto_datset['origin'].apply(origin)
```
